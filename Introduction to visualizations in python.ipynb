## Introduction to Data Visualization in Python
**Data visualization** *is a sub-displine of data analysis which allows us to understand data by placing it in visual context so as to explore patterns, trends and correlations that might not otherwise be detected or onbserved.*

There are over 20 common types visualization charts that one can use to tell the story in the data both in 2D and 3D. Python offers multiple great graphycal libraries that are packed with lots of different features to allow one to creat beautiful and informative charts. With python's visualization modules, one can easily create basic visualizations, interactive visualizations and even customize them into a dashboard.

The most common visualization packages in python are:
- [Matplotlib](https://matplotlib.org/): low level, provides lots of freedom
* [Pandas Visualization](https://pandas.pydata.org/pandas-docs/stable/visualization.html): easy to use interface, built on Matplotlib
* [Seaborn](https://seaborn.pydata.org/): high-level interface, great default styles
* [Plotly](https://plot.ly/python/): can create interactive plots

In this script, We shall go through how to create basic visualizations using the tools above. The session will generally focus on syntax and working with features provided in this visualization modules. The interpretation of charts will be covered in another session.

We shall use the free [Iris](https://archive.ics.uci.edu/ml/datasets/iris) and [Wine Reviews](https://www.kaggle.com/zynicide/wine-reviews) datasets in our practice.

## Matplotlib
This is the most popular library in python for making simple plots. Its a low-level library that offers freedom in defining your charts at the cost of having to write more codes. It has an intaface like that of Matlab.

#### Scatter Plots
To create a scatter plot in Matplotlib we can use the **scatter method**. We will also create a figure and an axis using **plt.subplots** so we can give our plot a title and labels.

```python
# create a figure and axis
fig, ax = plt.subplots(figsize=(12,6))

# scatter the sepal_length against the sepal_width
ax.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

we can make our chart more interesting by coloring the each data point by its species.

```python
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots(figsize=(12,6))
# plot each data-point
for i in range(len(iris['SepalLengthCm'])):
    ax.scatter(iris['SepalLengthCm'][i], iris['SepalWidthCm'][i],color=colors[iris['Species'][i]])
# set a title and labels
ax.set_title('Scatter Plot for Iris Dataset per species')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

## Line Chats
we can make lineplot in matplot lib using plot function.

```python
#get columns to plot
columns=iris.columns.drop(['Species'])

# Create x data
plot_data=range(0,len(iris))

# creating the figure and the axis
fig, ax = plt.subplots(figsize=(12,6))

## plot each column
for columns in columns:
    ax.plot(plot_data,iris[columns],label=columns)
    
#seting up the graph
ax.set_title('Line graph for Iris dataset using MatplotLib')
ax.legend()
```

#### Histogram 


```python
fig, ax = plt.subplots()
ax.hist(wine['points'])

ax.set_title('Wine Review Score')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

#### Bar Charts

```python
fig, ax = plt.subplots()
#Preparing the data to plot
plot_data=wine['points'].value_counts()

# Geting axis values
point=plot_data.index
frequency=plot_data.values

# Creating the graph
ax.bar(point,frequency)

# set title and labels 
ax.set_title('Wine Scores') 
ax.set_xlabel('Points') 
ax.set_ylabel('Frequency')
```

#### Bar Chart

#### Box Plots
A boxplot is a graphycal visualization of five statistics summary. i.e, minimum,25th%,median,75%,maximum .. and outliers

```python
plt.figure(figsize=(12,6))
plot_data = wine[(wine['points']>=95) & (wine['price']<1000)]
sns.boxplot('points', 'price', data=plot_data)
```

#### Heatmaps
A Heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. Heatmaps are perfect for exploring the correlation of features in a dataset.

```python
sns.heatmap(iris.corr(),annot=True)
```
